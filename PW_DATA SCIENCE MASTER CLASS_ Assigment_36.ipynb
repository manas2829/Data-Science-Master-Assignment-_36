## Q1. What is the different between a t-test and a z-test? Provide an example scenario where you would use each type of test.

### Ans:-

        A t-test and a z-test are both statistical hypothesis tests used to make inferences about population parameters 
        based on sample data. However, they differ in certain aspects, including the underlying assumptions and the 
        scenarios in which they are most appropriate.
      
          1. T-test:-
                  A t-test is used when the population standard deviation is unknown, and the sample size is relatively 
                  small (typically less than 30). It is based on the t-distribution, which has fatter tails than the 
                  standard normal (z) distribution.
                  
                  Example:- Suppose you want to compare the average scores of two groups of students (Group A and Group B)
                            in a math test. You randomly select a sample of 20 students from each group and record their
                            scores. In this case, you would use a t-test to determine if there is a significant difference
                            in the average scores between the two groups.
                  
          2. Z-test:-
                  A z-test is used when the population standard deviation is known, or when the sample size is large 
                  (typically greater than 30). It is based on the standard normal (z) distribution, which is symmetric
                  and has a mean of zero and a standard deviation of one.
                  
            Example :- Suppose you want to determine if the average height of a population of adults is significantly 
                       different from the average height reported in a national survey. You collect a large sample of 500
                       individuals and measure their heights. In this case, you would use a z-test because you have access
                       to the population standard deviation (from the national survey) and the sample size is large.
                       
    In summary, you would generally use a t-test when the population standard deviation is unknown and the sample size is 
    small, while a z-test is appropriate when the population standard deviation is known or when the sample size is large.               

## Q2. Differentiate between one-tailed and two-tailed tests.

### Ans:-

        One-tailed and two-tailed tests are terms used to describe the directionality of a statistical hypothesis test. 
        They refer to the regions of the probability distribution used to determine statistical significance. Here's the
        difference between the two:
        
        1.One-Tailed Test:-
                   In a one-tailed test, the statistical hypothesis is formulated to test for the difference or relationship
                   in a specific direction. The critical region is defined on only one side of the distribution, either the
                   right or the left side. This means that the test is sensitive to detecting effects in a particular 
                   direction.
                   
          Example :-  Suppose you want to investigate if a new drug improves cognitive performance. Your research hypothesis
                      is that the new drug will increase cognitive performance. In this case, you would use a one-tailed 
                      test and define the critical region on the right side of the distribution. The test would determine if
                      the mean cognitive performance of the drug group is significantly higher than the control group.
                      
      2. Two-Tailed Test:-
                   In a two-tailed test, the statistical hypothesis is formulated to test for the difference or relationship
                   in any direction. The critical region is divided equally on both sides of the distribution, capturing
                   extreme values in both tails. This means that the test is sensitive to detecting effects in either 
                   direction, whether it is an increase or decrease.  
                   
         Example :- Suppose you want to investigate if a teaching intervention affects student exam scores. Your research 
                    hypothesis is that the teaching intervention will lead to a significant difference in exam scores, but
                    you don't have a specific expectation about the direction of the difference. In this case, you would use
                    a two-tailed test, and the critical region would be divided on both sides of the distribution. The test
                    would determine if there is a significant difference in exam scores between the intervention group and 
                    the control group, regardless of whether the scores are higher or lower.
                    
        

## Q3. Explain the concept of Type 1 and Type 2 errors in hypothesis testing. Provide an example scenario for each type of error.

### Ans:-
            In hypothesis testing, Type 1 and Type 2 errors are two types of errors that can occur when making conclusions
            about a statistical hypothesis. Here's an explanation of each type and an example scenario for each:
            
        1.Type 1 Error(False Positive):-
             A Type 1 error occurs when we reject a null hypothesis that is actually true. In other words, we conclude 
             that there is a significant effect or relationship when, in reality, there is no such effect or relationship.
             It represents a false positive result.
             
       Example :- Suppose a pharmaceutical company is conducting clinical trials to test the effectiveness of a new drug 
                  for a specific medical condition. The null hypothesis states that the drug has no effect on the condition.
                  However, due to random chance or other factors, the sample data may lead to rejecting the null hypothesis,
                  concluding that the drug is effective when it is not. This would be a Type 1 error, as the drug is being
                  deemed effective when it actually has no effect.
                   
      2.Type 2 Error (False Nagative):-
            A Type 2 error occurs when we fail to reject a null hypothesis that is actually false. 
            In other words, we conclude that there is no significant effect or relationship when, in reality, 
            there is an effect or relationship. It represents a false negative result.
            
      Example :- Continuing from the previous example, suppose the null hypothesis is false, and the drug does have a real
                 effect on the medical condition. However, due to factors like small sample size or insufficient statistical
                 power, the sample data fails to provide enough evidence to reject the null hypothesis. As a result,the drug
                 is deemed ineffective when it actually has a positive effect. This would be a Type 2 error, as the drug is 
                 being incorrectly considered ineffective when it does have an effect.
                 
    In summary, a Type 1 error involves rejecting a true null hypothesis, leading to a false positive conclusion, while a 
    Type 2 error involves failing to reject a false null hypothesis, leading to a false negative conclusion. Both types of 
    errors are inherent risks in hypothesis testing and should be carefully considered when interpreting the results. The
    choice of significance level (alpha) and sample size can influence the likelihood of Type 1 and Type 2 errors.
    

## Q4.Explain Bayes's theorem with an example.

### Ans:-

      Bayes's theorem is a fundamental concept in probability theory that allows us to update our beliefs or probabilities 
      based on new evidence or information. It provides a way to calculate the conditional probability of an event given 
      prior knowledge or information about related events. Bayes's theorem is expressed mathematically as:
      
      P(A|B) = (P(B|A) * P(A)) / P(B)
      or
      P(B|A) = (P(A|B) * P(B) / P(A)
      
      Where :-
      
      P(A|B) is the probability of event A Occurring given event B has Occurred.
      P(B|A) is the probability of event B Occurring given event A has Occurred.
      P(A)   is the prior(Indepentent) probability of event A Occurring
      P(B)   is the prior(Indepentent) probability of event B Occurring
      
      Here is the Example to illustrate Bayes's theorem.
      
      Suppose there is a certain medical test for a rare disease. The test has been evaluated forits accuracy and following
      information is Know:-
          1. The probability of a person having a disease,P(A), is 0.01% ( 1% of the population has the disease)
          2. The probability of the test correctly detecting the disease,P(B|A), is 0.95(95% accuracy)
          3. The probability of a person not having the disease,P(~A), is 0.99 (1% of the population does not have the
             disease).
          4. The probability of the test incorrectly indicating the presence of the disease, P(B|~A), is 0.02 (2% false
             positive rate).
             
        Now, let's say a person takes the test, and it comes back positive (event B). We want to calculate the 
        probability that the person actually has the disease (event A|B).

        Using Bayes's theorem:

        P(A|B) = (P(B|A) * P(A)) / P(B)
        P(A|B) = (0.95 * 0.01) / P(B)

        To calculate P(B), we can use the law of total probability:
        
        P(B) = P(B|A) * P(A) + P(B|~A) * P(~A)
        P(B) = (0.95 * 0.01) + (0.02 * 0.99)
        
        Plugging these values into Bayes's theorem:

        P(A|B) = (0.95 * 0.01) / [(0.95 * 0.01) + (0.02 * 0.99)]

        After performing the calculation, let's assume we find P(A|B) to be 0.324 (32.4%).

        Therefore, based on the positive test result, the probability that the person actually has the disease is 0.324 
        or 32.4%.
      
                                  This example demonstrates how Bayes's theorem allows us to update the probability of an 
       event (having the disease) given new evidence (positive test result) in light of prior probabilities and conditional
       probabilities associated with the evidence and the event.

## Q5. What is the confidence interval ? How to calculated the confidence interval, Explain and Example.

### Ans:-

        A confidence interval is a range of values that is used to estimate an unknown population parameter, such as a 
        population mean or proportion. It provides a measure of the uncertainty associated with the estimate.
        
        To calculate a confidence interval, you typically follow these steps:
        
        1. Determine the confidence level:- The confidence level is a measure of how confident you want to be in your 
                                             estimate. It is usually expressed as a percentage, such as 90%, 95%, or 99%. 
                                             A higher confidence level corresponds to a wider interval.
                                             
        2. Collect a sample from the population:- Take a random sample from the population of interest. The sample should
                                                  be representative and sufficiently large to make valid inferences about 
                                                  the population.
                                                  
        3. Compute the sample statistic:- Calculate the sample statistic that estimates the population parameter you are 
                                          interested in. For example, if you want to estimate the population mean, you would
                                          calculate the sample mean.
                                          
        4. Determine the standard error:- The standard error is a measure of the variability of the sample statistic. It is
                                          calculated using the sample data and depends on the type of statistic being 
                                          estimated.
                                          
        5.Determine the critical value:- The critical value is based on the confidence level and the distribution of the 
                                         sample statistic. It is used to determine the margin of error around the sample
                                         statistic. Commonly used distributions include the normal distribution for large
                                         sample sizes or the t-distribution for small sample sizes.
                                         
        6.Calculate the margin of error: The margin of error represents the maximum amount by which the sample statistic is
                                         likely to differ from the population parameter. It is determined by multiplying the
                                         standard error by the critical value.
                                         
        7.Construct the confidence interval: Finally, you construct the confidence interval by taking the sample statistic 
                                             and adding/subtracting the margin of error. This creates a range of values 
                                             within which the population parameter is likely to fall with the specified
                                             confidence level.
                                             
        
        Example :- 
        
                    Suppose you want to estimate the average height of all adult males in a particular city. You take a 
            random sample of 100 adult males and measure their heights. The sample mean height is 175 cm, and the sample 
            standard deviation is 5 cm.You choose a 95% confidence level, which corresponds to a critical value of 1.96 for
            a large sample size (assuming a normal distribution).
            
            The standard error is calculated as the sample standard deviation divided by the square root of the sample size
            i.e: 5 cm / √100 = 0.5 cm.

            The margin of error is determined by multiplying the standard error by the critical value: 
            0.5 cm * 1.96 = 0.98 cm.
            
            Finally, the confidence interval is constructed by
            Upper level:- 175 + 0.98 = 175.98
            Lower level:- 175 - 0.98 = 174.02
            
            Conclusion:-Therefore, with 95% confidence, we estimate that the average height of all adult males in the city
                        falls within the range of 174.02 cm to 175.98 cm.

## Q7. Calculate the 95% confidence interval for a sample of data with a mean of 50  and a standard deviation of 5. Interpret the results.

### Ans:-

        To calculate the 95% confidence interval for a sample of data with a mean of 50 and a standard deviation of 5, 
        we need to consider the sample size. Without the sample size, it is not possible to calculate the standard error 
        and construct the confidence interval accurately. The sample size is a crucial component in estimating the precision 
        of the interval.

        Assuming you provide the sample size, let's say the sample size is n = 100.

        The formula to calculate the confidence interval for a population mean with a known standard deviation 
        (when the sample size is large) is:

        Confidence Interval = Sample Mean ± (Z * (Standard Deviation / √n))

        For a 95% confidence level, the Z-value is 1.96, which corresponds to the standard normal distribution.

        Using the given values, we can calculate the confidence interval:

        Confidence Interval = 50 ± (1.96 * (5 / √100))
                            = 50 ± (1.96 * 0.5)
                            = 50 ± 0.98
                            = (49.02, 50.98)

        Interpretation:
                        The 95% confidence interval for the population mean, based on the given sample data, is 
                        (49.02, 50.98). This means that we are 95% confident that the true population mean falls 
                        within this range.



```python
import scipy.stats as stats

# Sample mean and standard deviation
sample_mean = 50
sample_std = 5

# Sample size
sample_size = 100

# Calculate the standard error
standard_error = sample_std / (sample_size ** 0.5)

# Calculate the confidence interval
confidence_interval = stats.norm.interval(0.95, loc=sample_mean, scale=standard_error)

# Print the confidence interval
print("Confidence Interval:", confidence_interval)
```

## Q10. In study of effectiveness of a new weight loss drug, a sample of 50 participants lost an average of 6 pounds with standard deviation of 2.5 pounds. conduct a hypothesis test to determine if the drug is significantly effective at a 95% confidence level using a t-test.

### Ans:-

        To conduct a hypothesis test using a t-test, we need to set up the null hypothesis (H0) and the alternative 
        hypothesis (H1). In this case, we want to determine if the weight loss drug is significantly effective.

        Null hypothesis (H0): The weight loss drug is not significantly effective. (µ = 0)
        Alternative hypothesis (H1): The weight loss drug is significantly effective. (µ ≠ 0)

        We can perform a t-test to compare the sample mean to the hypothesized population mean (0 pounds) and determine if
        there is enough evidence to reject the null hypothesis.

        The formula for the t-score is:
        t = (x̄ - µ) / (s / √n)

                where:
                    x̄ is the sample mean,
                    µ is the hypothesized population mean,
                    s is the sample standard deviation, and
                    n is the sample size.

            In this case, the sample mean (x̄) is 6 pounds, the hypothesized population mean (µ) is 0 pounds, the sample 
         standard deviation (s) is 2.5 pounds, and the sample size (n) is 50 participants.

        Calculating the t-score:
            t   = (6 - 0) / (2.5 / √50)
                = 6 / (2.5 / 7.071)
                = 6 / 0.354
                = 16.949

            Next, we need to determine the critical value for a 95% confidence level. Since the sample size is 50,
            we will use a t-distribution with 49 degrees of freedom.

            The critical value for a two-tailed test at a 95% confidence level is approximately ±2.009 (using a t-table or 
            statistical software).

            Comparing the t-score to the critical value:
            Since the calculated t-score (16.949) is much larger than the critical value (±2.009), we can conclude that the 
            test statistic falls in the rejection region. Therefore, we reject the null hypothesis.

                    In conclusion, based on the results of the t-test, we have enough evidence to suggest that the weight 
            loss drug is significantly effective at a 95% confidence level.



```python
from scipy import stats
import numpy as np

# Sample data
sample_mean = 6
sample_std = 2.5
sample_size = 50

# Hypothesized population mean
hypothesized_mean = 0

# Conduct t-test
t_score, p_value = stats.ttest_1samp(np.random.normal(sample_mean, sample_std, sample_size), hypothesized_mean)

# Calculate the critical value for a 95% confidence level
critical_value = stats.t.ppf(0.975, df=sample_size-1)

# Compare t-score with the critical value and determine the result
if t_score > critical_value or t_score < -critical_value:
    result = "Reject the null hypothesis. The weight loss drug is significantly effective."
else:
    result = "Fail to reject the null hypothesis. The weight loss drug is not significantly effective."


print("t-score:", t_score)
print("p-value:", p_value)
print("Critical value:", critical_value)
print("Result:", result)
```

        The standard error (SE) is calculated using the formula:

        Standard Error = sqrt((p̂ * (1 - p̂)) / n)

                                where n is the sample size. In this case, the sample size is 500.

        Now we can plug in the values and calculate the confidence interval:

        Standard Error = sqrt((0.65 * (1 - 0.65)) / 500) ≈ 0.0188

        Margin of Error = 1.96 * 0.0188 ≈ 0.0368

        Confidence Interval = 0.65 ± 0.0368

        Therefore, the 95% confidence interval for the true proportion of people satisfied with their job is 
        approximately 0.6132 to 0.6868.
